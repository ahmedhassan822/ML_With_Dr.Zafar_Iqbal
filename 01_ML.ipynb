```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("/content/sample_data/california_housing_train.csv")
df
```

```python
plt.xlabel("total_rooms", color="red")
plt.ylabel("population",color="red")
plt.scatter(df.total_rooms,df.population)
#
```

```python
lr=linear_model.LinearRegression()
lr.fit(df[['total_rooms']],df.population)
```

```python
lr.predict([[25000]])
```

```python
lr.score(df[['total_rooms']],df.population)
```

```python
plt.plot(df.total_rooms,lr.predict(df[['total_rooms']]),color="red")
plt.scatter(df.total_rooms,df.population)
```
